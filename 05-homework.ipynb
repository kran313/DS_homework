# Normal
Вычислить с какой вероятностью пассажир женского пола, если известно, что он выжил (использовать теорему Байеса)

```python
# P(A|B) - надо найти по байесу

# P(B|A)
a = len(df[(df['Sex'] == 'female') & (df['Survived'] == 1)]) / len(df[df['Sex'] == 'female'])

# P(A) - женский пол
b = len(df[df['Sex'] == 'female']) / len(df)

# P(B) - выжил
c = len(df[df['Survived'] == 1]) / len(df)

a * b / c
```

# Normal+
Необходимо реализовать градиентный спуск с уменьшением  для функции двух переменных.

Проверить на функции 
. Начать из точки (5, 2).

Проверить свое решение на готовых функциях минимизации.

```python
def F(x, y):
    return (x - 3) ** 4 + (y + 5) ** 2
```

```python
def new(x, y):
    e = 0.001
    x1 = x - e * 4 * (x - 3) ** 3 if x > 3 else x + e * 4 * (x - 3) ** 3
    y1 = y - e * 2 * (y + 5) if y > -5 else y + e * 2 * (y + 5)
    
    return x1, y1
```

```python
x_start = 5
y_start = 2

for i in range(1000000):
    x_start, y_start = new(x_start, y_start)
```

```python
# ответ при маленьком шаге и большом количестве шагов

x_start, y_start
```

```python
# проверка

from scipy.optimize import minimize
x_start = [5, 2]
minimize(lambda x: F(x[0], x[1]), (x_start,))
```
