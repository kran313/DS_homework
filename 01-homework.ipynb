# Easy
Нужно написать функцию, которая на вход получает список из чисел, а на выходе дает список, где каждое число возведено в квадрат. Решать при помощи цикла.

```python
def squares_1(some_list):
    new_list = []
    for i in some_list:
        new_list.append(i ** 2)
        
    return new_list
```

```python
squares_1(list(range(10)))
```

```python
squares_1(list(range(100)))
```

```python
squares_1(list())
```

# Normal
Тоже самое, но на этот раз без цикла (смотрим раздел с функциями из ноутбука, который разбирали в классе). Сравнить скорость выполнения первого и второго варианта.

```python
def squares_2(some_list):
    return [i ** 2 for i in some_list]
```

```python
%%timeit
squares_1(list(range(10000)))
```

```python
%%timeit
squares_2(list(range(10000)))
```

# Hard
Написать функцию, которая на вход принимает список, а возвращает словарь со среднием, медианой и модой, например вот такой: {"mean": ..., "median": ..., "mode": ...}. На следующем занятии разберем как это сделать при помощи библиотек, а пока нужно реализовать просто по формулам.

```python
def func(some_list):
    mean, median, mode = 0, 0, [0, 0]
    
    for i in some_list:
        mean += i   
    mean /= len(some_list)
    
    
    new_list = sorted(some_list)
    if len(new_list) % 2 == 0:
        median = (new_list[len(new_list) - 1] + new_list[len(new_list)]) / 2
    else:
        median = new_list[int((len(new_list) - 1) / 2)]
        
    
    for i in set(some_list):
        if some_list.count(i) > mode[1]:
            mode = i, some_list.count(i)
        elif some_list.count(i) == mode[1]:
            mode = (mode[0] + i) / 2, some_list.count(i)
        
    
    return {'mean': float(f'{mean:.4f}'),
            'median': median,
            'mode': mode[0]}
```

```python
func([0, 1, 1, 10, 5, 4, 3])
```
